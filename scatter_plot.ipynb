```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_houses=pd.read_csv("https://raw.githubusercontent.com/data-doctors/kaggle-house-prices-advanced-regression-techniques/master/data/train.csv", delimiter=",",error_bad_lines=False)
```

```python
data_houses
```

```python
sns.lmplot(x='GrLivArea', y='SalePrice', data=data_houses)
```

```python
sns.lmplot(x='GrLivArea', y='SalePrice', data=data_houses, hue='OverallQual', fit_reg=False)
```

```python
sns.regplot(data_houses['GrLivArea'], data_houses['SalePrice'])
```

```python
sns.jointplot(data_houses['GrLivArea'], data_houses['SalePrice'])
```

```python
sns.jointplot(data_houses['GrLivArea'], data_houses['SalePrice'], kind='kde')
```
